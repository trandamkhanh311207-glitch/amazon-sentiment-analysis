```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/amazon_sentiment_processed.csv")
```

```python
df["sentiment"].value_counts().plot(kind="bar")
plt.title("Sentiment Distribution")
plt.show()
```

```python
df.groupby("sentiment")["review_length"].mean()
```

```python
df.groupby("sentiment")["vote"].mean()
```

```python
pd.crosstab(df["verified"], df["sentiment"], normalize="index")
```

## Key Insights from EDA
1. Negative reviews tend to be significantly longer than positive reviews, suggesting that dissatisfied customers provide more detailed feedback.
2. Negative reviews receive more helpful votes on average, indicating that they may contain more informative and useful content.
3. Verified purchases are more likely to be positive compared to non-verified ones, suggesting that actual buyers tend to have a better overall experiences.
These findings highlight meaningful behavioral patterns in user reviews and may inform feature engineering and modeling decisions.
